```julia
"""
Find the bifurcations on a 2D parameter grid
"""
function find_bifurcations_parameter_grid(df::DataFrame)
    # go through every row and draw a line between grid points if either number or stability of 
    # objects changes
    grid_length = Integer(sqrt(size(df,1)))
    dy = abs(df[1+grid_length,"grid_coordinates"][1]-df[1,"grid_coordinates"][1])
    dx = abs(df[2,"grid_coordinates"][2]-df[1,"grid_coordinates"][2])

    ddx = dx/10
    ddy = dy/10
    P=Plots.plot(size=(700,550),xtickfont=font(25),colorbar=true, legend = false,  font_family="sans-serif", ytickfont=font(25),xguidefontsize=32,yguidefontsize=32,legendfont=font(26),margin=7Plots.mm)
    heatmap!(P,x,y,log.(gradient_0011_noTF),c=:coolwarm,clims=(0,4.5),cscale=log,cgrad=(scale=:log))
    
    prog = Progress(nrow(df))
    for i = 1:nrow(df)
        #println(i)
        #display(P)
        if mod(i,grid_length)!=0
            if length.(df[i,"cycles"])!=length.(df[i+1,"cycles"])
                Plots.plot!(P,df[i,"grid_coordinates"][1]-dy/2:ddy:df[i,"grid_coordinates"][1]+dy/2,ones(length(df[i,"grid_coordinates"][1]-dy/2:ddy:df[i,"grid_coordinates"][1]+dy/2))*(df[i,"grid_coordinates"][2]+dx/2),legend=false,color="yellow",linewidth=5)
            else
                for n in eachindex(df[1,"cycles"])
                    same_stability_indices_found = []
                    for j in eachindex(df[i,"cycles"][n])
                        same_stability_indices = compare_stability(df[i,"eigvals"][n][j],df[i+1,"eigvals"][n])
                        # no match in stability means a bifurcation occured
                        if same_stability_indices === nothing
                            Plots.plot!(P,df[i,"grid_coordinates"][1]-dy/2:ddy:df[i,"grid_coordinates"][1]+dy/2,ones(length(df[i,"grid_coordinates"][1]-dy/2:ddy:df[i,"grid_coordinates"][1]+dy/2))*(df[i,"grid_coordinates"][2]+dx/2),legend=false,color="yellow
                            ",linewidth=5)
                            break
                        else
                            # if the cycle with same stability is also closest in state space -> no bifurcation
                            if length(same_stability_indices) == 1
                                #distance_cycles = get_minimal_state_space_distances(df[i,"cycles"][n][j],df[i+1,"cycles"][n])
                                #index_min = argmin(distance_cycles)
                                #if index_min != same_stability_indices[1] || same_stability_indices[1] in same_stability_indices_found
                                if same_stability_indices[1] in same_stability_indices_found
                                    Plots.plot!(P,df[i,"grid_coordinates"][1]-dy/2:ddy:df[i,"grid_coordinates"][1]+dy/2,ones(length(df[i,"grid_coordinates"][1]-dy/2:ddy:df[i,"grid_coordinates"][1]+dy/2))*(df[i,"grid_coordinates"][2]+dx/2),legend=false,color="yellow",linewidth=5)
                                else
                                    append!(same_stability_indices_found, same_stability_indices)
                                end
                            else
                                distance_cycles = get_combined_state_space_eigenvalue_distance(df[i,"cycles"][n][j],df[i+1,"cycles"][n], df[i,"eigvals"][n][j],df[i+1,"eigvals"][n])
                                index_min = argmin(distance_cycles)
                                if index_min in same_stability_indices_found
                                    Plots.plot!(P,df[i,"grid_coordinates"][1]-dy/2:ddy:df[i,"grid_coordinates"][1]+dy/2,ones(length(df[i,"grid_coordinates"][1]-dy/2:ddy:df[i,"grid_coordinates"][1]+dy/2))*(df[i,"grid_coordinates"][2]+dx/2),legend=false,color="yellow",linewidth=5)
                                else
                                    append!(same_stability_indices_found, index_min)
                                end
                            end
                        end
                    end
                end
            end
        end
        if i < size(df,1)-grid_length
            if length.(df[i,"cycles"])!=length.(df[i+grid_length,"cycles"])
                Plots.plot!(P,ones(length(df[i,"grid_coordinates"][2]-dx/2:ddx:df[i,"grid_coordinates"][2]+dx/2))*(df[i,"grid_coordinates"][1]+dy/2),df[i,"grid_coordinates"][2]-dx/2:ddx:df[i,"grid_coordinates"][2]+dx/2,legend=false,color="yellow",linewidth=5)
            else
                for n in eachindex(df[1,"cycles"])
                    same_stability_indices_found = []
                    for j in eachindex(df[i,"cycles"][n])
                        same_stability_indices = compare_stability(df[i,"eigvals"][n][j],df[i+grid_length,"eigvals"][n])
                        # no match in stability means a bifurcation occured
                        if same_stability_indices === nothing
                            Plots.plot!(P,ones(length(df[i,"grid_coordinates"][2]-dx/2:ddx:df[i,"grid_coordinates"][2]+dx/2))*(df[i,"grid_coordinates"][1]+dy/2),df[i,"grid_coordinates"][2]-dx/2:ddx:df[i,"grid_coordinates"][2]+dx/2,legend=false,color="yellow",linewidth=5)
                            break
                        else
                            # if the cycle with same stability is also closest in state space -> no bifurcation
                            if length(same_stability_indices) == 1
                                #distance_cycles = get_minimal_state_space_distances(df[i,"cycles"][n][j],df[i+grid_length,"cycles"][n])
                               # index_min = argmin(distance_cycles)
                                #if index_min != same_stability_indices[1] || same_stability_indices[1] in same_stability_indices_found
                                if same_stability_indices[1] in same_stability_indices_found
                                    Plots.plot!(P,ones(length(df[i,"grid_coordinates"][2]-dx/2:ddx:df[i,"grid_coordinates"][2]+dx/2))*(df[i,"grid_coordinates"][1]+dy/2),df[i,"grid_coordinates"][2]-dx/2:ddx:df[i,"grid_coordinates"][2]+dx/2,legend=false,color="yellow",linewidth=5)
                                else
                                    append!(same_stability_indices_found, same_stability_indices)
                                end
                            else
                                distance_cycles = get_combined_state_space_eigenvalue_distance(df[i,"cycles"][n][j],df[i+grid_length,"cycles"][n], df[i,"eigvals"][n][j],df[i+grid_length,"eigvals"][n])
                                index_min = argmin(distance_cycles)
                                if index_min in same_stability_indices_found
                                    Plots.plot!(P,ones(length(df[i,"grid_coordinates"][2]-dx/2:ddx:df[i,"grid_coordinates"][2]+dx/2))*(df[i,"grid_coordinates"][1]+dy/2),df[i,"grid_coordinates"][2]-dx/2:ddx:df[i,"grid_coordinates"][2]+dx/2,legend=false,color="yellow",linewidth=5)
                                else
                                    append!(same_stability_indices_found, index_min)
                                end
                            end
                        end
                    end 
                end
            end
        end
        next!(prog)
    end
    xlabel!(L"\Delta a_{00}")
    ylabel!(L"\Delta a_{11}")
    return P
end

data_path = "GT_models/GT_16/001"

data = load(data_path*"/GT16_0011.jld2","data")
model_path = "latent dimension/GT16/112"
#data = load(data_path*"/GT16_0011.jld2","data")
# parameters were loaded in python notebook visualization and stored as npy files
param_1 = npzread(model_path *"/param_1.npy")
param_2 = npzread(model_path *"/param_2.npy")


model_path = "latent dimension/GT16/111"
# parameters were loaded in python notebook visualization and stored as npy files
param_1_18 = npzread(model_path *"/param_1.npy")
param_2_18 = npzread(model_path *"/param_2.npy")


find_bifurcations_parameter_grid(data)
Plots.plot!(param_2[begin:160],param_1[begin:160],color="yellowgreen",linewidth=7)
Plots.plot!(param_2_18[begin:160],param_1_18[begin:160],color="yellowgreen",linewidth=7)
Plots.scatter!([param_2[42]],[param_1[42]],color="red",marker="o",markersize=6)

Plots.scatter!([param_2_18[56]],[param_1_18[56]],color="red",marker="o",markersize=6)
Plots.scatter!([param_2_18[101]],[param_1_18[101]],color="red",marker="o",markersize=6)

Plots.scatter!([param_2[1]],[param_1[1]],color="yellowgreen",marker=:star,markersize=10)

Plots.scatter!([param_2_18[1]],[param_1_18[1]],color="yellowgreen",marker=:star,markersize=10)
Plots.savefig(model_path*"/gradient_landscape_GT16.png")
k=Plots.scatter!([param_2_18[1]],[param_1_18[1]],color="yellowgreen",marker=:star,markersize=10)
display(k)
```
